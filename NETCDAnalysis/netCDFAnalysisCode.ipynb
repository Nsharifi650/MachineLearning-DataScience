```python
data = Dataset('SingleLevelData.dc','r')

# displaying the names of the variables
    
print(data.variables.keys())

# to look at each variable individually e.g. from line above 'lat' is one of the variables

lat = data.variables['lat']
print(lat)

time = data.variables['time']
print(time)

WindS = data.variables['wspeed']
print(WindS)

height = data.variables['height']
print(height)

#accessing the data from the variables
time_data = data.variables['time'][:]
windspeed = data.variables['wspeed'][:]
height_data = data.variables['height'][:]
Windspeed_inspection = data.variables['wspeed'][:,:,0,0]


# # quick plot
# plt.plot(time_data, windspeed[:,1,0,0])


xx = data.variables['height'][:]
yy = data.variables['wdir'][:]

Column_names2 = []

for i in range(xx.size):
    a = str(round((xx[i]),2))
    Column_names2.append('WS_'+a+'m')

for i in range(xx.size):
    a = str(round((xx[i]),2))
    Column_names2.append('WD_'+a+'_\xb0')
       
    
########## get the windspeed as a function of height and time

# #converting the time to date

starting_date = data.variables['time'].units[14:30] # 14:24string selection
TotalDays = time_data[time_data.size-1]/(60*24*60)

ending_date = pd.to_datetime(starting_date) + pd.DateOffset(days=TotalDays-1)

# # using pandas to create date range

date_range = pd.date_range(start = starting_date, end = ending_date,freq="60min").strftime('%Y-%m-%d %H:%M')
df = pd.DataFrame(0,columns = Column_names2,index = date_range)



Secsinhour = 60*60

index_time = np.zeros(date_range.size)

x = 0
Search_width = 10

######### updating loop
for i in range(date_range.size):
    hour = Secsinhour*(i+1)
    if time_data[0] > Secsinhour:
        index_time[0] = 0 
    for j in range(x,x+Search_width):
        if time_data[j] == hour:
            index_time[i] = j    
            x = j      
            

##getting the time averages now

averWinds = np.zeros((index_time.size,height_data.size))

for j in range(height_data.size):
    num = int(index_time[0])
    averWinds[0,j] = windspeed[num,j,0,0]
    for i in range(1,index_time.size):
        numb1 = int(index_time[i-1])
        numb2 = int(index_time[i])
        averWinds[i,j] = np.average(windspeed[numb1+1:numb2+1,j,0,0])
        
        
windDirection =  data.variables['wdir'][:]
    
windDirection =  data.variables['wdir'][:]
AngleRad = windDirection*math.pi/180
#Uspeed = np.dot(np.cos(AngleRad),windspeed)
Uspeed = np.cos(AngleRad)*windspeed
Vspeed = np.sin(AngleRad)*windspeed


UspeedAve= np.zeros((index_time.size,height_data.size))
VspeedAve= np.zeros((index_time.size,height_data.size))

for j in range(height_data.size):
    num = int(index_time[0])
    UspeedAve[0,j] = Uspeed[num,j]
    VspeedAve[0,j] = Vspeed[num,j]
    for i in range(1,index_time.size):
        numb1 = int(index_time[i-1])
        numb2 = int(index_time[i])
        UspeedAve[i,j] = np.average(Uspeed[numb1+1:numb2+1,j])
        VspeedAve[i,j] = np.average(Vspeed[numb1+1:numb2+1,j])


WindD = np.zeros((UspeedAve[:,1].size,height_data.size))
for i in range(UspeedAve[:,1].size):
    for j in range(Uspeed[1,:].size):
        WindD[i,j] = math.atan2(Vspeed[i,j], Uspeed[i,j])*180/math.pi
        if WindD[i,j] < 0:
            WindD[i,j] = WindD[i,j]+360
        

# #put the data in the df file

for jj in range(height.size):
    for ii in range(index_time.size):
        df.iloc[ii,jj] =averWinds[ii,jj]

for jj in range(height.size,height.size*2):
    nn = jj - height.size
    for ii in range(index_time.size):
        df.iloc[ii,jj] =WindD[ii,nn]        
        
df = df.rename_axis('Date/ Time')
df.to_csv('Dowa wind speed per hour average plus wind direction average - optimised.csv')



lat_value = data.variables['lat'][:]
lon_value = data.variables['lon'][:]

with open('LongitudinalAndLatitudeValues.txt', 'w') as f:
    f.write('The latitudinal and longitudinal values are ' 
            + str(lat_value) + ',' +str(lon_value))
```
